**For Loop:**

```python
# Q.1 Python program to print all even numbers between 1 to 100 using a for loop.
for i in range(1, 101):
    if i % 2 == 0:
        print(i, end=" ")
```

```python
# Q.2 Python program to print all odd numbers between 1 to 100 using a for loop.
for i in range(1, 101):
    if i % 2 != 0:
        print(i, end=" ")
```

```python
# Q.3 Python program to print sum of all even numbers between 1 to 100 using a for loop.
sum_even = 0
for i in range(1, 101):
    if i % 2 == 0:
        sum_even += i
print("Sum of even numbers =", sum_even)
```

```python
# Q.4 Python program to print sum of all odd numbers between 1 to 100 using a for loop.
sum_odd = 0
for i in range(1, 101):
    if i % 2 != 0:
        sum_odd += i
print("Sum of odd numbers =", sum_odd)
```

```python
# Q.5 Python program to count the digits of a given number using for loop.
num = int(input("Enter a number: "))
count = 0
for digit in str(num):
    count += 1
print("Total digits =", count)
```

**While Loop:**


```python
# Q.1 Python program to print all even numbers between 1 to 100 using a while loop.
i = 1
while i <= 100:
    if i % 2 == 0:
        print(i)
    i += 1
```

```python
# Q.2 Python program to print all odd numbers between 1 to 100 using a while loop.
i = 1
while i <= 100:
    if i % 2 != 0:
        print(i)
    i += 1
```

```python
# Q.3 Python program to print sum of all even numbers between 1 to 100 using a while loop.
i = 1
sum_even = 0
while i <= 100:
    if i % 2 == 0:
        sum_even += i
    i += 1
print("Sum of even numbers =", sum_even)
```

```python
# Q.4 Python program to print sum of all odd numbers between 1 to 100 using a while loop.
i = 1
sum_odd = 0
while i <= 100:
    if i % 2 != 0:
        sum_odd += i
    i += 1
print("Sum of odd numbers =", sum_odd)
```

```python
# Q.5 Python program to count the digits of a given number using while loop.
num = int(input("Enter a number: "))
count = 0
temp = num
while temp > 0:
    temp = temp // 10
    count += 1
print("Total digits =", count)
```
